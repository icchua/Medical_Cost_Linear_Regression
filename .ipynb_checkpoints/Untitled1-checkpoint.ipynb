```python
values = dt_graph.sex
v_counts = values.value_counts()
total = v_counts.sum()
def my_fmt(x):
    # Calculate the actual percentage value from x (fraction)
    percentage = x / 100.0 * total
    return '{:.4f}%\n({:.0f})'.format(x, percentage)
```

# Checking the outliers

We need to find out if there are any outliers and we need to exclude them in our dataset

```python
# check the percentile and median base distribution(For visualising the outliers)
fig= plt.figure(figsize=(18,10))

ax=fig.add_subplot(2,2,1)
sns.boxplot(data=dt, x=dt["age"], y=dt["charges"], hue=None, color='r', ax=ax)
ax.set_title("Charges vs Age ")

ax=fig.add_subplot(2,2,2)
sns.boxplot(data=dt, x=dt["bmi"], y=dt["charges"], hue=None, color='b', ax=ax)
ax.set_title("Charges vs BMI")

ax=fig.add_subplot(2,2,3)
sns.boxplot(data=dt, x=dt["children"], y=dt["charges"], hue=None, color='c', ax=ax)
ax.set_title("Charges vs children")


plt.savefig(".\images\outliers.png", dpi = 150, bbox_inches = 'tight')
plt.show()
```

```python
# We will check the children, bmi and age if there are any outliers
columns= ['children', 'bmi', 'age']
outliers=[]

for col in columns:
    #print(col)
    mean=np.mean(dt[col])
    std=np.std(dt[col])
    for i in range(len(dt)):
        item = dt[col].iloc[i]
        z_score=(item-mean)/std
        if np.abs(z_score)>3:
            outliers.append(i)

outliers = set(outliers)
outliers = list(outliers)
```

```python
dt.shape
```

```python
# Now we can drop the outliers
dt.drop(dt.index[outliers],inplace=True)
```

```python
# Now let us check the visuals without the outliers
fig= plt.figure(figsize=(18,10))

ax=fig.add_subplot(2,2,1)
sns.boxplot(data=dt, x=dt["age"], y=dt["charges"], hue=None, color='r', ax=ax)
ax.set_title("Charges vs Age ")

ax=fig.add_subplot(2,2,2)
sns.boxplot(data=dt, x=dt["bmi"], y=dt["charges"], hue=None, color='b', ax=ax)
ax.set_title("Charges vs BMI")

ax=fig.add_subplot(2,2,3)
sns.boxplot(data=dt, x=dt["children"], y=dt["charges"], hue=None, color='c', ax=ax)
ax.set_title("Charges vs children")


plt.savefig(".\images\outliers2.png", dpi = 150, bbox_inches = 'tight')
plt.show()
```

Now let us transform our data to fit in our needs

```python
# We need to change the data to be more usable to do that we will assign sex, smoker and region numerical values
# sex - 0 for female and 1 or male
# smoker - 0 for no and 1 for yes
# region - 1 for northeast, 2 for northwest, 3 for southeast and 4 for southwest

dt['sex'].replace({'female': 0, 'male': 1}, inplace=True)
dt['smoker'].replace({'no': 0, 'yes' : 1}, inplace=True)
dt['region'].replace({'northeast': 1, 'northwest': 2, 'southeast': 3, 'southwest': 4}, inplace=True)
```

```python
dt_pred
```

```python
# Library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(data):
    vif = pd.DataFrame()
    vif['features'] = data.columns
    vif['VIF_Values'] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
    
    return vif
```

```python
# Using the VIF to check if there are high correlations with the variables
calculate_vif(dt_pred)
```

```python
# We should make a histogram and see to confirm that the data we have is categorical
fig, axs = plt.subplots(2, 3, figsize=(8, 8))

columns = ['sex', 'children','smoker','region','age', 'bmi']

for i, column in enumerate(columns):
    row = i // 3
    col = i % 3
    dt.plot(kind='scatter', x=column, y='charges', ax=axs[row, col], alpha=0.4, color='b')
    axs[row, col].set_xlabel(column)
    axs[row, col].set_ylabel('Frequency')

plt.tight_layout()
plt.savefig(".\images\categorical.png", dpi = 150, bbox_inches = 'tight')

plt.show()
```

so we need to make use dummy coding to so we can make use of then if our linear regression.

```python
# first we should bin the age so that it would be easier for us to read the data
# and it might help in reducing multicollinearity
bins = [17, 22, 26, 30, 34, 38, 42, 46, 50, 54, 58, 62, 66]
labels = ['18_22','23_26','27_30','31_34','35_38','39_42','43_46','47_50','51_54','55_58','59_62', '63_66']
```

```python
dt['age_group'] = pd.cut(dt['age'], bins=bins, labels=labels)
dt.drop("age", axis = 1, inplace = True)
```

```python
categoricals = ['age_group', 'sex', 'children', 'smoker', 'region']
```

```python
age_dummies = pd.get_dummies(dt.age_group, prefix  = "ages", drop_first = True)
sex_dummies = pd.get_dummies(dt.sex, prefix  = "sex", drop_first = True)
child_dummies = pd.get_dummies(dt.children, prefix  = "child", drop_first = True)
smoker_dummies = pd.get_dummies(dt.smoker, prefix  = "smoker", drop_first = True)
region_dummies = pd.get_dummies(dt.region, prefix  = "region", drop_first = True)
```

```python
dummies_full = pd.concat([age_dummies, sex_dummies, child_dummies, smoker_dummies, region_dummies], axis = 1)
```

```python
#now we can drop all the categorical data as it would hinder us in checking if the data needs to be transformed and normalized
dt.drop(['sex', 'children', 'smoker', 'region', 'age_group'], axis = 1, inplace = True)
```

We can see that there is no multicollinearty that we need to address.

# Validation

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error , r2_score
```

```python
dt_last = dt_3.copy()
```

```python
# I believe we need to transform our data so that it would fit our model better
dt_last.charges = np.log(dt_last.charges)
```

```python
y = dt_last['charges']
X = dt_last.drop('charges', axis = 1)
```

### Train - Test Split

```python
# Let us now test it at random_state 4
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
#y_pred = linreg.predict(X_test)
LinearRegression()
```

### Calculating the Mean Squared Error

```python
y_hat_train = linreg.predict(X_train)
y_hat_test = linreg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

train_mse = mean_squared_error(y_train, y_hat_train)
test_mse = mean_squared_error(y_test, y_hat_test)
print('Train Mean Squared Error:', train_mse)
print('Test Mean Squared Error: ', test_mse)
```

```python
train_mse - test_mse
```

### Model Validation Comments

The difference is little at 0.03 or 3% difference between its Mean Squared error. I believe that the model is at its best fit.

### R-Squared Value of the Model

```python
print('R2 score: {:.2%}'.format(r2_score(y_test, y_hat_test)))
```

We can see here that the r-squared of our model is quite high at 0.7914 which is better than out model 3 r-squared at 0.7580

```python
# Get top 5 variables

from sklearn.feature_selection import RFE

selector = RFE(linreg, n_features_to_select=9)
selector = selector.fit(X_train, y_train.values)

selector.support_
```

```python
#Get top 5 Contributors for a high medical charges
selected_columns = X_train.columns[selector.support_]
linreg.fit(X_train[selected_columns], y_train)
print(selected_columns)
```
