```python
# We would put everything except for reno_bool and waterfront as we would need a data that we can use to check if renovations can raise
# the prices
columns= ['children', 'bmi', 'age']
outliers=[]

for col in columns:
    #print(col)
    mean=np.mean(dt[col])
    std=np.std(dt[col])
    for i in range(len(dt)):
        item = dt[col].iloc[i]
        z_score=(item-mean)/std
        if np.abs(z_score)>3:
            outliers.append(i)

outliers = set(outliers)
outliers = list(outliers)
```

```python
# Now we can drop the outliers so we have a more neater dataset
dt.drop(dt.index[outliers],inplace=True)
```

```python
# We need to change the data to be more usable to do that we will assign sex, smoker and region numerical values
# sex - 0 for female and 1 or male
# smoker - 0 for no and 1 for yes
# region - 1 for northeast, 2 for northwest, 3 for southeast and 4 for southwest

dt['sex'].replace({'female': 0, 'male': 1}, inplace=True)
dt['smoker'].replace({'no': 0, 'yes' : 1}, inplace=True)
dt['region'].replace({'northeast': 1, 'northwest': 2, 'southeast': 3, 'southwest': 4}, inplace=True)
```

```python
# We need to address the data type so we can use them
dt.sex = dt.sex.astype(int)
dt.smoker = dt.smoker.astype(int)
dt.region = dt.region.astype(int)
```
